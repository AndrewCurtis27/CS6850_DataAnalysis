```python
class MyListener(Stream):
    
    def __init__(self, time_limit=30):
        self.start_time = time.time()
        self.limit = time_limit
        self.outFile = open('output.json', 'a')
        super(MyListener, self).__init__()
        
    def on_data(self, data):
        if (time.time() - self.start_time) < self.limit:
            self.outFile.write(data)
            self.outFile.write('\n')
            return True
        else:
            self.outFile.close()
            return False
    
    def on_error(self, status):
        print(status)
```

```python
tweets_friends = api.get_friends(screen_name='huhi')
tweets_followers = api.get_followers(screen_name='huhi')

columns = ['tweet_id', 'name',  'screen_name', 'location',
            'description', 'followers_count', 'friends_count', 
            'favorites_count', 'creation_time', 'number_of_tweets', 
            'verified', 'follower_followee']
data = []

for tweet in tweets_friends:
    data.append([tweet.id, tweet.name, tweet.screen_name, tweet.location,
                tweet.description, tweet.followers_count, tweet.friends_count, 
                tweet.favourites_count, tweet.created_at, tweet.statuses_count, 
                tweet.verified, 'followee'])
    
for tweet in tweets_followers:
    data.append([tweet.id, tweet.name, tweet.screen_name, tweet.location,
                tweet.description, tweet.followers_count, tweet.friends_count, 
                tweet.favourites_count, tweet.created_at, tweet.statuses_count, 
                tweet.verified, 'follower'])

df = pd.DataFrame(data, columns = columns)
```
