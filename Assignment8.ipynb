```python
import pandas as pd
from sklearn import cluster
```

# a)

```python
data = pd.read_csv("traffic.csv")
```

```python
data
```

# b)

```python
data = data.dropna()
```

```python
data.columns
```

```python
data = data.drop(['date'], axis=1)
data
```

```python
Y = data['class']
pd.unique(Y)
```

```python
data = data.drop(['class'], axis=1)
data
```

```python
# Y.replace(['none', 'afternoon', 'evening'], [0,1,2])
Y = Y.replace({'none':0, 'afternoon':1, 'evening':2})
Y
```

# c)

```python
k_means = cluster.KMeans(n_clusters=3)
k_means.fit(data)
centroids = k_means.cluster_centers_
labels = k_means.labels_
```

```python
print(centroids)
print(labels)
```

# d)

## confusion matrix

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(Y, labels)
```

## Adjusted random index (ARI)

```python
from sklearn.metrics.cluster import adjusted_rand_score

adjusted_rand_score(Y, labels)
```

## Adjusted Mutual Information

```python
from sklearn.metrics.cluster import adjusted_mutual_info_score
adjusted_mutual_info_score(Y, labels)
```

## Homogeneity

```python
from sklearn.metrics.cluster import homogeneity_score
homogeneity_score(Y, labels)
```

## Completeness

```python
from sklearn.metrics.cluster import completeness_score
completeness_score(Y, labels)
```

## V-measure

```python
from sklearn.metrics.cluster import v_measure_score
v_measure_score(Y, labels)
```

## Silhouette score

```python
from sklearn import metrics
metrics.silhouette_score(data, labels)
```

## Davies-Bouldin index

```python
metrics.davies_bouldin_score(data, labels)
```
